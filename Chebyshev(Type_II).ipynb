```
##Import required library
from scipy import signal
from scipy.fft import fft
import numpy as np
import matplotlib.pyplot as plt
```

```
##Given specification
N=17          ##Order of the filter
fp=np.array([50,200])        ##Pass band edge frequencies in Hz
Fs=2000         ##Sampling frequency in Hz
```

```
##Determine the filter coefficients in SOS form using signal.iirfilter function
sos = signal.iirfilter(N,fp,rp=None,rs=70,btype='band',analog=False,ftype='cheby2',fs=2000,output='sos')
```

```
##Compute frequency response using signal.sosfreqz function
w, h =signal.sosfreqz(sos,worN=2000,fs=2000)
```

```
##Plot the filter magnitude response
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)
ax.semilogx(w, 20 * np.log10(np.maximum(abs(h), 1e-5)), 'r', linewidth=2)
ax.set_title('Chebyshev Type II bandpass frequency response', fontsize=20)
ax.set_xlabel('Frequency [Hz]', fontsize=20)
ax.set_ylabel('Amplitude [dB]', fontsize=20)
ax.axis((10, 1000, -100, 10))
ax.grid(which='both', axis='both')
plt.show()
```
